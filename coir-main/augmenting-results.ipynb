```python
import pandas as pd
```

```python
def augment(res_df_path, output_path):
    expl_path = '/work/pi_wenlongzhao_umass_edu/27/anamikaghosh/CoSQA_explanations_query_code.csv'
    expl_df = pd.read_csv(expl_path)
    expl_df.rename(columns={"query_id": "query-id", "corpus_id": "corpus-id"}, inplace=True)
    expl_df = expl_df[['query-id', 'doc', 'corpus-id', 'code','explanation_deepseek_1_cleaned']]
    
    res_df = pd.read_csv(res_df_path)
    
    merged_df = res_df.merge(
        expl_df[['corpus-id', 'doc', 'code', 'explanation_deepseek_1_cleaned']],
        left_on='retrieved_doc_id',
        right_on='corpus-id',
        how='left'
    )
    merged_df.rename(columns={"doc": "retrieved_doc","code": "retrieved_code", "explanation_deepseek_1_cleaned":"retrieved_explanation"}, inplace=True)
    merged_df = merged_df.merge(
        expl_df[['query-id', 'doc']],
        left_on='query_id',
        right_on='query-id',
        how='left'
    )
    merged_df.rename(columns={"doc": "query"}, inplace=True)
    
    merged_df = merged_df[['query_id', 'query','retrieved_doc_id', 'retrieved_doc', 'retrieved_code', 'retrieved_explanation','score', 'ground_truth_relevance']]
    
    merged_df.to_csv(output_path+'augmented_retrieval_analysis.csv', index=False)
```

```python
expl_path = '/work/pi_wenlongzhao_umass_edu/27/anamikaghosh/CoSQA_explanations_query_code.csv'
expl_df = pd.read_csv(expl_path)
expl_df.rename(columns={"query_id": "query-id", "corpus_id": "corpus-id"}, inplace=True)
expl_df = expl_df[['query-id', 'doc', 'corpus-id', 'code','explanation_deepseek_1_cleaned']]
expl_df.head()
```

```python
res_df = pd.read_csv('/work/pi_wenlongzhao_umass_edu/27/vaishnavisha/CS696DS-Oracle-Retrieving-Code-Explanations/coir-main/results/intfloat/e5-base-v2_explanation_deepseek_1/retrieval_evaluation.csv')
res_df.head()
```

```python
merged_df = res_df.merge(
    expl_df[['corpus-id', 'doc', 'code', 'explanation_deepseek_1_cleaned']],
    left_on='retrieved_doc_id',
    right_on='corpus-id',
    how='left'
)
merged_df.rename(columns={"doc": "retrieved_doc","code": "retrieved_code", "explanation_deepseek_1_cleaned":"retrieved_explanation"}, inplace=True)
merged_df = merged_df.merge(
    expl_df[['query-id', 'doc']],
    left_on='query_id',
    right_on='query-id',
    how='left'
)
merged_df.rename(columns={"doc": "query"}, inplace=True)

merged_df.head()
```

```python
merged_df = merged_df[['query_id', 'query','retrieved_doc_id', 'retrieved_doc', 'retrieved_code', 'retrieved_explanation','score', 'ground_truth_relevance']]
```

```python
merged_df.to_csv('/work/pi_wenlongzhao_umass_edu/27/vaishnavisha/CS696DS-Oracle-Retrieving-Code-Explanations/coir-main/results/intfloat/e5-base-v2_explanation_deepseek_1/augmented_retrieval_analysis.csv', index=False)
```

```python
final_results = []

for query_id, group in merged_df.groupby('query_id'):
    top_row = group.loc[group['score'].idxmax()]
    
    ground_truth_rows = group[group['ground_truth_relevance'] == 1]
    if not ground_truth_rows.empty:
        ground_truth_row = ground_truth_rows.iloc[0]
        final_results.append(top_row)
        final_results.append(ground_truth_row)

final_df = pd.DataFrame(final_results)
final_df.head()
```
